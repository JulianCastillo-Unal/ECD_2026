````python
# @markdown # **Funciones**
# @markdown
# @markdown Doble clic en mí y podrás ver mi código.
# =====================================================================================================================
# REGION: Funciones Útiles para Taller Proyecto Integrador
# NOTA: Las funciones fueron comentadas con GitHub Copilot
# =====================================================================================================================
def ImprimirLogActual(ruta:str):
    """
    Esta función abre un archivo en la ruta proporcionada y luego imprime su contenido.

    Parámetros:
    ruta (str): La ruta al archivo que se va a abrir y leer.

    """
    with open(ruta, 'r') as archivo:
        print(archivo.read())

def IdentificarLogs(directorio:str):
    """
    Esta función recorre todos los archivos en un directorio proporcionado,
    identifica los archivos con la extensión .log y luego imprime la ruta completa hacia estos archivos.

    Parámetros:
    directorio (str): El directorio que se va a recorrer para buscar archivos .log.

    """
    for archivo in os.listdir(directorio):
        if archivo.endswith(".log"):
            # Construir la ruta completa hacia el archivo
            ruta_completa = os.path.join(directorio, archivo)
            console.print(ruta_completa, style="url")

def ClasificadorAcceso(dato):
    """
    Esta función verifica si un dato es una cadena de texto y si contiene la frase 'Open Access'.

    Parámetros:
    dato (str): La cadena de texto a verificar.

    Devuelve:
    True si el dato es una cadena de texto y contiene 'Open Access', False en caso contrario.
    """
    if isinstance(dato, str):
        if 'Open Access' in dato:
            return True
        else:
            return False
    else:
        return False

def ContarAutores(dato):
    """
    Esta función cuenta el número de elementos en una lista.

    Parámetros:
    dato (list): La lista cuyos elementos se contarán.

    Devuelve:
    El número de elementos en la lista si el dato es una lista, 0 en caso contrario.
    """
    if isinstance(dato, list):
        return len(dato)
    else:
        return 0

def ImpresionDiccionario(data):
    """
    Imprime un diccionario en una tabla estilizada utilizando un tema personalizado con rich.

    Args:
        data (dict): Diccionario con los datos a imprimir. Las claves y valores serán estilizados.

    Returns:
        None: Solo imprime la tabla estilizada en la consola.
    """
    # Crear la tabla
    table = Table(title="Estilos Rich")

    # Agregar columnas con estilos
    table.add_column("Clave", style="normalb", no_wrap=True)
    table.add_column("Valor", style="normalb")

    # Rellenar la tabla con los datos del diccionario
    for key, value in data.items():
        table.add_row(str(key), str(value))  # Convertir valores a str

    # Imprimir la tabla
    console.print(table)
````

## **Transformaciones de Datos**

*   Separar los nombres de autores: La columna "AUTORES" contiene múltiples autores separados por punto y coma. Vamos a separar estos nombres en una lista.
*   Convertir años a datos numéricos: Asegurarnos de que la columna "ANIO" esté en formato numérico.
*   Extracción y limpieza de palabras clave: Unir las columnas de palabras clave en una sola lista para análisis.
*   Análisis de acceso abierto: Limpiar y categorizar los datos de acceso abierto.

¿Se pueden realizar otro tipo de transformaciones o ajustes a los datos?


````python
data.info()
````

````python
#Separar los autores
data['LISTAUTORES'] = data['AUTORES'].str.split('; ')
# Convertir la columna 'ANIO' a formato numérico
data['ANIO'] = pd.to_numeric(data['ANIO'], errors='coerce')
print('Se valida el uso de año numerico al dataframe')
# Unir las columnas de palabras clave en una sola lista para análisis
data['KEYWORDS'] = data['PCLAVEA'].fillna('') + '; ' + data['PCLAVEI'].fillna('')
print('Se unen las columnas de palabras clave en una sola')
data['ALLKEYWORDS'] = data['KEYWORDS'].str.split('; ')
print('Se crea la columna ALLKEYWORDS donde se almacenan todas las palabras clave')
# Limpiar y categorizar los datos de acceso abierto

data['OPENACCESS'] = data['ACCESO'].apply(ClasificadorAcceso)
data['CANTIDADAUTORES'] = data['LISTAUTORES'].apply(ContarAutores)
print('Se aplican las funciones ClasificadorAcceso al campo ACCESO y ContarAutores al campo LISTAUTORES')
# Ver los primeros datos transformados
data.head(2)
````

## **Frecuencia de Autores**

Explode, Transforma cada elemento de una lista en una fila, replicando los valores del índice.

````python
# Contar la frecuencia de cada autor
print('Calculo y grafica de la frecuencia de autores.')
autores = data['LISTAUTORES'].explode()
cuentauores = Counter(autores)
print(f'El total de filas es {len(autores)}')
````

````python
CantidadAutores = 25 # @param {type:"slider", min:5, max:50, step:5}
````

````python
#Seleccionamos la cantidad de autores a visualizar
top_autores = cuentauores.most_common(CantidadAutores)

# Convertir a un DataFrame para facilitar la visualización
top_autores_df = pd.DataFrame(top_autores, columns=['Author', 'Count'])

# Filter out rows where 'Author' is NaN
top_autores_df = top_autores_df.dropna(subset=['Author'])

# Crear la figura y el gráfico de barras horizontales
plt.figure(figsize=(10, 6))
bars = plt.barh(top_autores_df['Author'], top_autores_df['Count'], color='skyblue')
# Agregar etiquetas con los valores al final de cada barra
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2,
             f'{width}', ha='left', va='center')
# Configurar etiquetas y título
plt.xlabel('Número de Publicaciones')
plt.ylabel('Autores')
plt.title(f'Top {CantidadAutores} Autores más Frecuentes')
# Invertir el eje Y para que se muestre de mayor a menor
plt.gca().invert_yaxis()
plt.grid('on')
plt.show()
````

````python
with Progress() as progress:
    # Crear una barra de progreso
    task = progress.add_task("[green]Procesando datos atipicos de ALLKEYWORDS", total=len(data))
    contadorborrado = 0
    for index, row in data.iterrows():
        allkeys = row['ALLKEYWORDS']
        #Validaciones sobre el vector
        if '' in allkeys:
            allkeys.remove('')
        if ' ' in allkeys:
            allkeys.remove(' ')
        if len(allkeys)== 0:
            contadorborrado+=1
            data.drop(index, inplace=True)
        progress.update(task, advance=1, total=len(data))
    progress.update(task, total=len(data), completed=len(data), description="Completado")
````

## **Autores más citados**

````python
# Assuming 'data_concitas' DataFrame has 'CITACIONES' and 'LISTAUTORES' columns
# Merge with data_concitas to get author list for each publication with more than 5 citations
impacto_autores = data_concitas[['TITULO', 'CITACIONES', 'LISTAUTORES']].merge(data_concitas[['TITULO', 'CITACIONES']], on=['TITULO', 'CITACIONES'], how='left')

# Explode the LISTAUTORES column to have one author per row
impacto_autores_exploded = impacto_autores.explode('LISTAUTORES')

# Group by author and sum the citations
cited_authors = impacto_autores_exploded.groupby('LISTAUTORES')['CITACIONES'].sum().sort_values(ascending=False)

# Select the top N authors based on the topublicaciones slider
top_cited_authors = cited_authors.head(topublicaciones)

# Create a bar plot for the top cited authors
plt.figure(figsize=(10, 6))
bars = plt.barh(top_cited_authors.index, top_cited_authors.values, color='teal')

# Add labels to the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2,
             f'{width}', ha='left', va='center')

plt.xlabel('Total Citaciones')
plt.ylabel('Autor')
plt.title(f'Top {topublicaciones} Autores Más Citados')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest count at the top
plt.grid(True)
plt.show()
````
